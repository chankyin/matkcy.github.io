# Vector Fields: Macroscopic and Microscopic Circulation

Consider the following vector fields:


$\displaystyle {\bf F} = <-y,x>$




$\displaystyle {\bf G} = < \frac{-y}{x^{2}+y^{2}},~~\frac{x}{x^{2}+y^{2}}>$

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

matplotlib.rcParams['xtick.direction'] = 'out'
matplotlib.rcParams['ytick.direction'] = 'out'
```

# Macroscopic Circulation

```python
delta = 0.5
x = np.arange(-5.0, 5.0, delta)
y = np.arange(-5.0, 5.0, delta)
X, Y = np.meshgrid(x, y)
#Z=f(X,Y)

levels = arange(1, 10, 1.0)

U = -Y/(X**2+Y**2)
V = X/(X**2+Y**2)

P=-Y
Q=X

plt.figure(1)
plt.streamplot(X, Y, P, Q, color=P, linewidth=2)
plt.title('Vector field F and G')

plt.figure(2)
plt.streamplot(X, Y, U, V, color=U, linewidth=2)
```

## Discussion:  Place a free-floating object in both of the vector fields. By looking at the fields lines, it is not difficult to believe that this object will rotate in a circular path about the origin. This is called macroscopic circulation.  

# Microscopic Circulation  $~~~~$

```python
delta = 0.5
x = np.arange(10.0, 20.0, delta)
y = np.arange(10.0, 20.0, delta)
X, Y = np.meshgrid(x, y)

levels = arange(1, 10, 1.0)

U = -Y/(X**2+Y**2)
V = X/(X**2+Y**2)

P=-Y
Q=X

plt.figure(1)
plt.streamplot(X, Y, P, Q, color=P, linewidth=2)
plt.title('Vector field F')

plt.figure(2)
plt.streamplot(X, Y, U, V, color=U, linewidth=2)
plt.title('Vector field G ')
```

## Discussion: In the above plots of the vector field F and G, we have retricted to a smaller domain for better view. At first glance, both plots look almost the same. A closer examination shows that the colors change differently across the region for these vector fields. The blue lines correspond to larger values, whereas the red ones give smaller values. 

##  Now, imagine placing a small disk with its centre fixed at a point, say at (15,15). Notice that unlike a free-floating object, the disk can only rotate about its centre. What can we say about this rotation for the vector field F and G? 

## For vector field F:

## Consider the vectors acting on the boundary of the disk near the top and bottom. These vectors are generally pointing in the north-west direction. However, their magnitudes are different: those near the top seem to have larger magnitude than those near the bottom. Consequently, the disk will rotate anticlockwise.

## For vector field G:

## Consider the vectors acting on the boundary of the disk near the top and bottom. These vectors are generally pointing in the north-west direction. Unlike the previous case, their magnitudes also seem to be more or less the same! Hence, the disk does not seem to have a tendency to rotate about its centre. In fact, it will not rotate at all.

# Conclusion:

## Macroscopic and microscopic circulations are different. It is possible that a vector field has macroscopic circulation but not microsciopic circulation. The latter is also known as curl, a concept which we will examine later on in the course.
